```python
import os
import cv2
from ultralytics import YOLO

# Đường dẫn dự án và mô hình
PROJECT_DIR = os.path.abspath(os.path.join(os.getcwd(), '..'))
model_path = os.path.join(PROJECT_DIR, '../models/motorbike-tracking/train5/weights/best.pt')

# Khởi tạo mô hình YOLO
model = YOLO(model_path)

# Thêm mã để sử dụng webcam
cap = cv2.VideoCapture(1) 

while True:
    ret, frame = cap.read()  # Đọc khung hình từ webcam
    if not ret:
        break

    results = model(frame)  # Dự đoán trên khung hình
    annotated_frame = results[0].plot()  # Vẽ kết quả lên khung hình

    cv2.imshow('Webcam Test', annotated_frame)  # Hiển thị khung hình đã xử lý

    if cv2.waitKey(1) & 0xFF == ord('q'):  # Nhấn 'q' để thoát
        break

cap.release() 
cv2.destroyAllWindows()  


```
