```python
import os
import cv2
from ultralytics import YOLO

# ng dn d n v m hnh
PROJECT_DIR = os.path.abspath(os.path.join(os.getcwd(), '..'))
model_path = os.path.join(PROJECT_DIR, '../models/motorbike-tracking/train5/weights/best.pt')

# Khi to m hnh YOLO
model = YOLO(model_path)

# Thm m  s dng webcam
cap = cv2.VideoCapture(1) 

while True:
    ret, frame = cap.read()  # c khung hnh t webcam
    if not ret:
        break

    results = model(frame)  # D on trn khung hnh
    annotated_frame = results[0].plot()  # V kt qu ln khung hnh

    cv2.imshow('Webcam Test', annotated_frame)  # Hin th khung hnh  x l

    if cv2.waitKey(1) & 0xFF == ord('q'):  # Nhn 'q'  thot
        break

cap.release() 
cv2.destroyAllWindows()  


```
